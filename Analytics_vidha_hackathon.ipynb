```python
df4.Gender=df4.Gender.apply(lambda x:1 if x=="Male" else 0)
df4.Married=df4.Married.apply(lambda x:1 if x=="Yes" else 0)
df4.Education=df4.Education.apply(lambda x:1 if x=="Graduate" else 0)
df4.Self_Employed=df4.Self_Employed.apply(lambda x:1 if x=="Yes" else 0)
df4.Property_Area=df4.Property_Area.apply(lambda x:1 if x=="Urban" else 0)
df4.Loan_Status=df4.Loan_Status.apply(lambda x:1 if x=="Y" else 0)
```

# Model Building 

```python
import statsmodels.api as sm 
```

```python
logm1=sm.GLM(y_train,(sm.add_constant(X_train)), family=sm.families.Binomial())
logm1.fit().summary()
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
```

```python
from sklearn.feature_selection import RFE
rfe=RFE(logreg,10)
rfe=rfe.fit(X_train,y_train)
```

```python
rfe.support_
```

```python
list(zip(X_train,rfe.support_,rfe.ranking_))
```

```python
col=X_train.columns[rfe.support_]
```

```python
X_train.columns[~rfe.support_]
```

```python
col
```

```python
X_train=sm.add_constant(X_train[col])
logm2=sm.GLM(y_train,X_train,family=sm.families.Binomial())
res=logm2.fit()
res.summary()
```

```python
y_train_pred=res.predict(X_train)
y_train_pred[:10]

```

```python
y_train_pred=y_train_pred.values.reshape(-1)
y_train_pred[:10]
```

```python
y_train_pred_final=pd.DataFrame({'Loan':y_train.values,'Loan_prob':y_train_pred,'Load_ID':df4['Loan_ID']})
y_train_pred_final
```

```python
# creatinmg new column 

y_train_pred_final['predict']=y_train_pred_final.Loan_prob.map(lambda x:1 if x>0.5 else 0)
y_train_pred_final
```

```python
from sklearn import metrics
```

```python
confusion= metrics.confusion_matrix(y_train_pred_final.Loan,y_train_pred_final.predict)
confusion
```

```python
print(metrics.accuracy_score(y_train_pred_final.Loan,y_train_pred_final.predict))
```
