# Data Preparation

## Create Dummy Variables for all categorical features

```python
dum1=pd.get_dummies(df['season'], drop_first=True)
df=pd.concat([df,dum1],axis=1)

#creating dummy variables for season as it is not binary 
```

```python
dum2=pd.get_dummies(df['mnth'], drop_first=True)
df=pd.concat([df,dum2],axis=1)

#creating dummy variables for mnth as it is not binary
```

```python
dum3=pd.get_dummies(df['weekday'], drop_first=True)
df=pd.concat([df,dum3],axis=1)

#creating dummy variables for weekday as it is not binary
```

```python
dum4=pd.get_dummies(df['weathersit'], drop_first=True)
df=pd.concat([df,dum4],axis=1)

#creating dummy variables for weathersit as it is not binary
```

```python
df.drop(['season','mnth','weekday','weathersit'],axis=1,inplace=True)

#As dummy variables are created 
```

```python
df.head(10)
```

## Divide the Data to train and test

```python
from sklearn.model_selection import train_test_split
np.random.seed(0)
df_train, df_test=train_test_split(df,train_size=0.7,test_size=0.3,random_state=100)
```

```python
df_train
```

## Perform Scaling 

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()

#Scaling the numerical variables to reduce the difference between coefficients
```

```python
num_vars=['temp','hum','windspeed','cnt']
df_train[num_vars]=scaler.fit_transform(df_train[num_vars])
```

```python
df_train.head()
```

## Dividing the Data into X and y 

```python
y_train=df_train.pop('cnt')
X_train=df_train
```

# Data Modelling and Evaluation

## Creating Linear Regression using Mixed Approach

### Selecting 15 variables Using RFE Technique

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Creating X_test dataframe with RFE selected variables


lm=LinearRegression()
lm.fit(X_train,y_train)
rfe=RFE(lm,15)
rfe.fit(X_train,y_train)
```

```python
list(zip(X_train.columns,rfe.ranking_,rfe.support_))
```

```python
col=X_train.columns[rfe.support_]
col
```

```python
X_train_rfe=X_train[col]
```

### Using Manual Method





```python
#Creating Linear Regression Model with 15 variables

import statsmodels.api as sm
X_train_rfe =sm.add_constant(X_train_rfe)
lr=sm.OLS(y_train,X_train_rfe).fit()
lr.params
```

```python
lr.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
#Checking VIF for all variables

vif=pd.DataFrame()
vif['features']=X_train_rfe.columns
vif['VIF']=[variance_inflation_factor(X_train_rfe.values,i) for i in range(X_train_rfe.shape[1])]
vif['VIF']=round(vif['VIF'],2)
vif=vif.sort_values(by='VIF',ascending=False)
vif
```
